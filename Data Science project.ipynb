#### Training data

```python
from sklearn.model_selection import train_test_split

# Feature variables
x = df['num_of_months']
y = df['cargo_throughput']


X = np.asarray(x).reshape(-1,1)
Y = np.asarray(y).reshape(-1,1)

# Splitting the data, creating the train and test sets
Xtrain,Xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.3 ,random_state=0)
```

```python
print(Xtrain.shape, Xtest.shape, ytrain.shape, ytest.shape)
```

```python
from sklearn import linear_model

# fit a linear model
lm = linear_model.LinearRegression()
l_model = lm.fit(Xtrain, ytrain)
l_predicted = lm.predict(Xtest)
```

```python
# Predicted value of my cargo throughput 
l_predicted
```

## Conclusions 

### Performance of results 

```python
from sklearn.metrics import r2_score,mean_squared_error, accuracy_score

accuracy= r2_score(ytest, l_predicted)          
# I am using r2score to test the accuracy of my training model and rounding up the result to 2 decimal place
print("Accuracy of the regression model above is : ",round(accuracy*100,2), "%")

# Linear regression stats
l_train_score = l_model.score(Xtrain, ytrain)
l_test_score = l_model.score(Xtest, ytest)
l_MSE = mean_squared_error(ytest, l_predicted)
# Root Mean Squared Error
l_root_MSE = np.sqrt(l_MSE) 
# r2 square 
l_r2_square = np.sqrt(r2_score(ytest, l_predicted))



print('Train Score: {}'.format(l_train_score))
print('Test Score: {}'.format(l_test_score))
print('Mean Squared Error: {}'.format(l_MSE))
print('Root Mean Squared Error: {}'.format(l_root_MSE))
print('R2 Square: {}'.format(l_r2_square))
```
